```python
from PIL import Image, ImageDraw 
from PIL import Image, ImageFilter   
  
# creating a image object 

im1 = Image.open(r"cosmos.jpg") 

for i in range(1):
    im2=im1.filter(ImageFilter.RankFilter(size = 5, rank =3 * 3-1))
    im2.show()
```

```python
from PIL import Image 
import PIL 
  
# creating a image object (main image) 
im1 = Image.open(r"maa.jpg") 
  
# rotating a image 180 deg counter clockwise 
im1 = im1.rotate(180, PIL.Image.NEAREST, expand = 1) 
  
# to show specified image 
im1.show() 
```

```python
from PIL import Image, ImageDraw 
from PIL import Image, ImageChops  
  
# creating a image1 object 

im1 = Image.open(r"flag.jpg") 
im2 = Image.open(r"ful.jpg")
 
im3 = ImageChops.darker(im1, im2) 
    
im3.show()  
```

```python
from PIL import Image 
import PIL 
  
# creating a image object (main image) 
im = Image.open(r"maa.jpg") 
  
im1 = Image.Image.split(im) 
  
# showing each band 
x=im1[0].show() 
y=im1[1].show() 
z=im1[2].show() 


im1[0].save('first.jpg')
im1[1].save('second.jpg')
im1[2].save('third.jpg')
```

```python
# Import modules
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Read images
img1 = mpimg.imread('first.jpg')
img2 = mpimg.imread('second.jpg')
img3 = mpimg.imread('third.jpg')
img4 = mpimg.imread('ful.jpg')

# Set canvas (fig) and the number of axes (images) I want to display
#I want to display my 4 images in a grid 2x2

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(15,10))

ax1.imshow(img1)
ax1.set_title("Image 1")
 
ax2.imshow(img2)
ax2.set_title("Image 2")
ax3.imshow(img3)
ax3.set_title("Image 3")
ax4.imshow(img4)
ax4.set_title("Image 4")

plt.show()

```

```python
# The same example with just the first two images in a grid 1x2

fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize=(15,10))

ax1.imshow(img1)
ax1.set_title("Image 1")
ax2.imshow(img2)
ax2.set_title("Image 2")
```
