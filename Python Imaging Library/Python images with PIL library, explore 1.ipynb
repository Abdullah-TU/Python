```python
# importing image object from PIL 
import math 
from PIL import Image, ImageDraw 
  
w, h = 220, 190
shape = [(10, 10), (w - 10, h - 10)] 
  

img = Image.new("RGB", (w, h)) 
img = Image.eval(img, lambda x: x/2)

for i in range(1, 5):
    img1 = ImageDraw.Draw(img)   
    img1.rectangle(shape, fill ="white", outline ="red")    
    
    img.show()    
```

```python
from PIL import Image, ImageDraw 
from PIL import Image, ImageOps 
  
# creating a image1 object 
im1 = Image.open(r"draw.jpg") 
  
# applying fit method 
# Setting width = 100 and height = 100 

for i in range(4):
    im2 = ImageOps.fit(im1, (100, 100), method = 0, bleed = 0.0, centering =(0.8, 0.8))
    im3 = Image.eval(im2, (lambda x: 552 - x * 15)) 
    im3.show()
```
