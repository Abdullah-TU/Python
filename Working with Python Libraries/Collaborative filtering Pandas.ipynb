```python

import pandas as pd 
data=pd.read_csv('lastfm-matrix-germany.csv', index_col='user')
```

```python

import numpy as np
xin=np.count_nonzero(data, axis=1)<10
inds=data.index.where(xin).dropna()
```

```python
yin=np.count_nonzero(data, axis=0)<35
cols=data.columns.where(yin).dropna()

data=data.drop(index=inds, columns=cols)
data

```

```python

from scipy.spatial.distance import pdist,  squareform 
artdists=squareform(pdist(data.T, "cosine"))
artdists[artdists==0]=1 
print('distance: ', end='')
printed=[]  
for i, j in enumerate(artdists.argmin(axis=0)):
    if (i,j) not in printed: 
        print('({},{})'.format(data.columns[i],data.columns[j]), end=' ')
        printed.append((j,i))

```

```python
usrdists=squareform(pdist(data, "cosine"))
usrdists[usrdists==0]=1
print('distance: ', end='')
printedUsrs=[]  # For avoiding printing same users again
for i, j in enumerate(usrdists.argmin(axis=0)):
    if (i,j) not in printed: 
        print('({},{})'.format(data.index[i],data.index[j]), end=' ')
        printedUsrs.append((j,i))
```

```python
artdistsDF=pd.DataFrame(artdists, index=data.columns, columns=data.columns)
artists=['rihanna', 'nirvana', 'kanye west']
for name in artists:
    print(name+':', end=' ')
    for val,sim in zip(artdistsDF[name].sort_values()[0:10].values, artdistsDF[name].sort_values()[0:10].index):
        print('({:.4f},{})'.format(val, sim), end=' ')
    print('\n')

```
